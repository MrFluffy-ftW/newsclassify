```python
url = "https://kathmandupost.com//"
response = requests.get(url)
soup = BS(response.content, "html.parser")
```

### function for cleaning scrapped data

```python
CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext

def clean_text(text):
  clean_str = ""
  text_str = str(text)
  text_str = cleanhtml(text_str)
  text_str = text_str.replace("\xa0","")
  text_str = text_str.replace("\u202f","")
  text_str = text_str.replace("[","")
  text_str = text_str.replace("\'s","'s")

  clean_str += " "
  clean_str += text_str  
  return clean_str
```

### extracting links from website

```python
actual_links = []

main_container = soup.find("main")
web_links = main_container.find_all("a")

for links in web_links:
    links = links['href']
    if links.find("https://kathmandupost.com/") == -1:
        links = "https://kathmandupost.com" + links
    actual_links.append(links)

#removing duplicate links
actual_links = list(dict.fromkeys(actual_links))

actual_links
```

### Extracting information


```python
paragraph = []
date = []
title = []
news_link = []
count = 0
for link in actual_links:
    try:
        url = link
        response = requests.get(url)
        news_link.append(link)
        soup = BS(response.content, "html.parser")
        heading = soup.find(attrs={"style": "margin-bottom:0.1rem;"}).get_text()
        temp_container=soup.find("section",class_='story-section')
        paragraph_temp = temp_container.find_all('p')
        clean_temp = clean_text(paragraph_temp)
        clean_temp = clean_temp.replace('.,','.')
        print(type(clean_temp))
        
        paragraph.append(clean_temp)
        title.append(heading)
        time = soup.find("div",class_='blocktop blocktop-date').get_text()
        date.append(time)
        news_link.append(link)
        if count > 2:
            break
        count += 1
        # print(time)
        # print(paragraph)
    except:
        continue
    
```
