```python
import requests
from bs4 import BeautifulSoup as BS
import re
```

```python
#url
url = "https://www.setopati.com/"
response = requests.get(url)
soup = BS(response.content, "html.parser")
```

```python
CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
def clean_text(text):
  clean_str = ""
  text_str = str(text)
  text_str = cleanhtml(text_str)
  text_str = text_str.replace("\xa0","")
  text_str = text_str.replace("\u202f","")
  text_str = text_str.replace("[","")
  text_str = text_str.replace("\'s","'s")
  clean_str += " "
  clean_str += text_str  
  return clean_str
```

```python
'''Extracting links'''

actual_links = []
main_container = soup.find("div",id='content')
link = main_container.find_all("a")
# print(link)
temp_web_links = [web_link['href'] for web_link in link] 

#removing faltu links
for links in temp_web_links:
    if links.find("https://www.setopati.com") == 0:
        actual_links.append(links)

#removing duplicate links
actual_links = list(dict.fromkeys(actual_links))
print(actual_links)
```

```python
paragraph = []
date = []
title = []
news_link = []
#extracting news articles
for link in actual_links:
    try:
        url = link
        response = requests.get(url)
        soup = BS(response.content, "html.parser")
        heading = soup.find("h1",class_='news-big-title').getText()
        print(heading,"\n")
        news_body = soup.find("div",class_='editor-box')
        temp = news_body.find_all("p")
        temp_clean = clean_text(temp)
        paragraph.append(temp_clean)
        title.append(heading)
        time = soup.find("span",class_='main-title').getText()
        date.append(time)
        news_link.append(link)
    except:
        continue
```

```python
print(len(news_link))
```

```python


import pandas as pd

df = pd.DataFrame()
for i in range(len(paragraph)):
    df = df.append({"title" : title[i],
                    "date" : date[i],
                    "news" : paragraph[i],
                    "news_link": news_link[i]},ignore_index = True)

df.to_csv("setopati.csv",index=False)
```
